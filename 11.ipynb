### 11. Faça uma análise dos filmes e mostre para o investidor uma análise gerencial dos filmes no dataset (análise livre para mostrar sua criatividade, desenvolvimento e habilidades do Python).

```python
import pandas as pd
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/title_names.tsv', sep='\t')
```

```python
#Importando arquivo:
ratings = pd.read_csv('Arquivos/t_ratings.tsv',sep='\t')
```

```python
#Fazendo a junção das tabelas:
juncao = pd.merge(title_names,ratings, on= 'tconst', how='left')
```

```python
#Filtrando para aparecer somente a categoria de filmes:
juncao = juncao[juncao['titleType']=='movie']
```

```python
#Vizualizar resultado com os 60 primeiros:
juncao.head(60)
```

```python
#Analisando a relação de anos com quantidade de filmes estreados:
juncao['startYear'].value_counts().head(60)
```

```python
#Fazendo um tratamento na coluna genres para separar os gêneros:
juncao['genres'] = juncao.apply(lambda x: x['genres'].replace(',',' '), axis=1)
```

```python
#Separando os gêneros em colunas distintas:
sep = juncao['genres'].str.split(expand=True)

#Visualizar resultado:
sep
```

```python
#Criando tres colunas para os valores do separador, em junção:
juncao[['1','2','3']]= sep
```

```python
#Visualizar resultado:
juncao
```

```python
#Criando uma tabela para cada coluna de gênero junto com as outras colunas:
g1=juncao[['tconst','primaryTitle','titleType','startYear','runtimeMinutes','genres','averageRating','numVotes','1']]
g2=juncao[['tconst','primaryTitle','titleType','startYear','runtimeMinutes','genres','averageRating','numVotes','2']]
g3=juncao[['tconst','primaryTitle','titleType','startYear','runtimeMinutes','genres','averageRating','numVotes','3']]
```

```python
#Renomeando as colunas responsáveis pelos gêneros para o mesmo nome:
g1.rename(columns={'1':'genero'},inplace=True)
g2.rename(columns={'2':'genero'},inplace=True)
g3.rename(columns={'3':'genero'},inplace=True)
```

```python
#Fazendo o empilhamento das 3 tabelas:
union = pd.concat([g1,g2,g3])
```

```python
#Visualizar resultado:
union
```

```python
#Agrupando pelo gênero e tirando a média das avaliações e numero de votos de acordo com cada gênero:
media_gen_filmes = union.groupby('genero')[['averageRating','numVotes']].mean().copy().reset_index()
```
