### 6. Qual o filme com maior tempo de duração? Calcule os percentis considerando o tempo de duração. 

```python
import pandas as pd
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/title_names.tsv', sep='\t')
```

```python
#Selecionando apenas as colunas que quero presentes na tabela:
title_names = title_names[['tconst','titleType','primaryTitle','runtimeMinutes','genres']]
```

```python
#Filtrando a tabela para aparecer somente a categoria de "movie":
title_names = title_names[title_names['titleType']=='movie']
```

```python
#Vizualizar resultado:
title_names
```

```python
#Criando nova coluna com os minutos e tratando os valores "\N" por 0:
title_names['Duration'] = title_names.apply(lambda x: x['runtimeMinutes'].replace(r'\N','0'),axis=1)
```

```python
#Transfomando a coluna criada para int:
title_names = title_names.astype({'Duration':'int'})

#Vizualizar resultado:
title_names
```

```python
#Criando nova coluna chamada "genero" com os valores pré-tratados:
title_names['genero'] = title_names.apply(lambda x: x['genres'].replace(',',' '),axis=1)
```

```python
#Separando os gêneros em colunas diferentes:
sep = title_names['genero'].str.split(expand=True)

#Vizualizar resultado:
sep
```

```python
#Incluindo coluna 'Duration' na tabela sep:
sep['Duration'] = title_names.apply(lambda x: x['Duration'],axis=1)
```

```python
#Incluindo a coluna "primaryTitle" na ta bela sep:
sep['primaryTitle'] = title_names.apply(lambda x: x['primaryTitle'],axis=1)
```

```python
#Vizualizar resultado:
sep
```

```python
#Renomeando o nome das colunas:
sep.columns = ['g1','g2','g3','Duration','primaryTitle']

#Vizualizar resultado:
sep
```

```python
#Separando as colunas em tabelas distintas, para depois fazer o empilhamento:
g1 = sep[['g1','Duration','primaryTitle']].copy()
g2 = sep[['g2','Duration','primaryTitle']].copy()
g3 = sep[['g3','Duration','primaryTitle']].copy()

#Vizualizar resultado:
display(g1,g2,g3)
```

```python
#Renomeando o nome da coluna responsável pelos generos e deixando o nome em todas as tabelas iguais:
g1.rename(columns={'g1':'genero'},inplace=True)
g2.rename(columns={'g2':'genero'},inplace=True)
g3.rename(columns={'g3':'genero'},inplace=True)
```

```python
#Empilhando as tabelas:
union = pd.concat([g1,g2,g3])

#Vizualizar resultado:
union
```

```python
#Identificando os gêneros, para posteriormente filtrar apenas os desejados:
union['genero'].drop_duplicates()
```

```python
#Excluindo os gêneros que não são considerados gêneros de filme:
tratamento = union[~union['genero'].isin([r'\N', 'Documentary','Game-Show','News','Reality-TV','Short',
                                              'Talk-Show',None])]
```

```python
#Resetando o index:
tratamento = tratamento.reset_index(drop=True)
```

```python
#Ordenando os valores em ordem decrescente de Duração:
maxDuration = tratamento.sort_values(by='Duration', ascending = False)
```

```python
#Vizualizar resultado:
maxDuration
```

```python
#Filtrando para aparecer somente minutos diferentes de 0:
maxDuration = maxDuration[maxDuration['Duration']!=0]

#Vizualizar resultado:
maxDuration.head(5)
```

```python
#Chamando a funcção describe() para identificar os percentis e algumas outras infomações: 
maxDuration.describe()
```
