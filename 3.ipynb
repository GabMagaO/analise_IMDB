### 3. Qual a mediana de avaliação dos filmes Por Gênero?

```python
import pandas as pd
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/t_basics.tsv', sep ='\t')
```

```python
#Filtrando para aparecer na tabela, apenas os filmes:
title_names = title_names[title_names['titleType'] == 'movie']

#Visualizar resultado:
title_names
```

```python
#Importando arquivo:
avaliações = pd.read_csv('Arquivos/t_ratings.tsv', sep='\t')
```

```python
#Visualizar resultado:
avaliações
```

```python
#Fazer uma junção das tabelas a partir de tconst:
juncao = pd.merge(title_names, avaliações, on=['tconst'],how='left')

#Selecionar apenas as colunas que eu quero mostrar em minha tabela:
juncao = juncao[['tconst','titleType','primaryTitle','genres','averageRating','numVotes']]

#Visualizar resultado:
juncao
```

```python
#Criando uma coluna nova aplicando o tratamento:
juncao['genTratados'] = juncao.apply(lambda x: x['genres'].replace(',',' '),axis=1)

#Resetando o index:
juncao= juncao.reset_index(drop=True)

#Visualizar resultado:
juncao
```

```python
# Eu separei, em colunas distintas, os Generos Tratados:
separador=juncao['genTratados'].str.split(expand=True)

#Resetando o index:
separador = separador.reset_index(drop=True)

#Visualizar resultado:
separador
```

```python
#Renomeando as colunas:
separador.columns= ['g1','g2','g3']

#Visualizar resultado:
separador
```

```python
#Crio uma variável com os valores de tconst nela:
tconst=juncao['tconst'] 

#Crio uma coluna nova, em meu separador, com os valores de tconst:
separador['tconst']=tconst

#Visualizar resultado:
separador
```

```python
#Crio uma variável para cada coluna de genero:
g1=separador[['g1','tconst']].copy()
g2=separador[['g2','tconst']].copy()
g3=separador[['g3','tconst']].copy()

#Visualizar resultado:
display(g1)
display(g2)
display(g3)
```

```python
#Renomeio as colunas de cada tabela, deixando todos os nomes iguais:
g1.rename(columns={'g1':'genero'},inplace=True)
g2.rename(columns={'g2':'genero'},inplace=True)
g3.rename(columns={'g3':'genero'},inplace=True)
```

```python
#Em seguida empilho essas colunas com nomes iguais:
union = pd.concat([g1,g2,g3])

#Visualizar resultado:
union
```

```python
#Para visualizar se o código funcionou, eu chamo o id do filme que antes era um gênero e passou a ser três:
union[union['tconst']=='tt0000574' ]
```

```python
#Faço a junção dessa nova tabela "union" com a tabela "juncao":
merge=pd.merge(union,juncao,on=['tconst'],how='left')
```

```python
#Seleciono apenas as colunas que eu quero a minha tabela apresente:
merge = merge[['tconst','titleType','genero','primaryTitle','averageRating','numVotes']]

#Visualizar resultado:
merge
```

```python
#Agrupo os generos pegando a mediana das avaliações:
merge=merge.groupby(by='genero')[['averageRating']].median().copy()
```

```python
#Reseto index:
merge=merge.reset_index()

#Visualizar resultado:
merge
```

```python
#Em seguida eu crio uma variável nova , que vai ser a resposta, excluindo os gêneros que não são filmes:
avaliacao_mediana_genero = merge[~merge['genero'].isin(['Talk-Show','Game-Show','Short','Biography',
                                                        'Talk-Show','Reality-TV','News','Documentary'])]

```

```python
#Visualizar resultado:
avaliacao_mediana_genero
```

```python
#E por fim, eu excluo a linha de índice 28, que corresponde a nulo:
avaliacao_mediana_genero=avaliacao_mediana_genero.drop(index=28)
```
