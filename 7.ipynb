### 7. Qual a relação entre duração e gênero?

```python
import pandas as pd
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/title_names.tsv',sep='\t')
```

```python
#Filtrando tabela para apresentar apenas filmes:
movies = title_names[title_names['titleType']=='movie']
```

```python
#Selecionando apenas as colunas que quero que apresente em minha tabela:
movies = movies[['tconst','titleType','primaryTitle','runtimeMinutes','genres']]
```

```python
#Criando coluna chamada 'Duration' com os valores de 'runtimeMinutes', alterando "\N" por 0:
movies['Duration'] = movies.apply(lambda x: x['runtimeMinutes'].replace(r'\N','0'),axis=1)
```

```python
#Vizualizar resultado:
movies
```

```python
#Tranfomando a coluna Duration para int:
movies = movies.astype({'Duration':'int'})

#Vizualizar resultado:
movies.dtypes
```

```python
#Criando uma nova coluna com os generos pré-tratados:
movies['genTratados'] = movies.apply(lambda x: x['genres'].replace(',',' '),axis=1)

#Vizualizar resultado:
movies
```

```python
#Seprando os generos em colunas distintas:
sep = movies['genTratados'].str.split(expand=True)

#Criando uma nova coluna "Duration" com os valores da "Duration" da tabela movies:
sep['Duration'] = movies.apply(lambda x: x['Duration'],axis=1)

#Vizualizar resultado:
sep
```

```python
#Renomeando as colunas:
sep.columns= ['g1','g2','g3','Duration']

#Vizualizar resultado:
sep
```

```python
#Criando uma nova coluna "primaryTitle" com os valores de 'primaryTitle' da tabela movies:
sep['primaryTitle'] = movies.apply(lambda x: x['primaryTitle'],axis=1)

#Vizualizar resultado:
sep
```

```python
#Criando uma tabela para cada coluna de gênero:
g1 = sep[['g1','Duration','primaryTitle']].copy()
g2 = sep[['g2','Duration','primaryTitle']].copy()
g3 = sep[['g3','Duration','primaryTitle']].copy()

#Vizualizar resultado:
display(g1,g2,g3)
```

```python
#Renomeando o nome da coluna responsável pelos generos e deixando o nome em todas as tabelas iguais:
g1.rename(columns={'g1':'generos'},inplace=True)
g2.rename(columns={'g2':'generos'},inplace=True)
g3.rename(columns={'g3':'generos'},inplace=True)
```

```python
#Empilhando as tablas e resetando o index:
union = pd.concat([g1,g2,g3]).reset_index(drop=True)

#Vizualizar resultado:
union
```

```python
#Analisando os gêneros apresentados na tabela:
union['generos'].drop_duplicates()
```

```python
#Retirando os gêneros que, não são considerados gêneros de filmes, da tabela:
union = union[~union['generos'].isin([r'\N','Documentary','Short','News','Talk-Show','Reality-TV','Game-Show',
                                     None])]
```

```python
#Filtrando para aparecer apenas filmes com Duration diferente de 0:
union = union[union['Duration']!=0]

#Vizualizar resultado:
union
```

```python
#Fazendo a contagem de quantas vezes, filmes de mesmo gênero, possuem quantidade de minutos iguais:
union[['generos','Duration']].value_counts().head(60)

#Uma boa parte dos gêneros de filmes analisados, tem maior quantidade de filmes com 90 min.
#Filmes de generos: Drama, Comedy, Action, Thriller, Romance, Crime, Horror. Possuem maior quantidade de filmes com 90 min.
```

```python
#Retiro os generos apresentados no output de cima:
union2=union[~union['generos'].isin(['Drama','Comedy','Action','Romance','Thriller','Crime','Horror'])]

#E realmente confirmo que a maioria dos gêneros te mais filmes com 90 min. 
union2[['generos','Duration']].value_counts().head(60)
```
