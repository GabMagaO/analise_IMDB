### 4. Qual a mediana de avaliação dos filmes em relação ao ano de estreia?

```python
import pandas as pd
from datetime import datetime,date
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/title_names.tsv',sep='\t')
```

```python
#Visualizar resultado:
title_names
```

```python
#Importando arquivo:
avaliacoes = pd.read_csv('Arquivos/t_ratings.tsv', sep ='\t')
```

```python
#Visualizar resultado:
avaliacoes
```

```python
#Faço ma junção das tabelas:
juncao = pd.merge(title_names,avaliacoes, on=['tconst'],how='left')

#Seleciono apenas as colunas que eu quero presentes na minha tabela:
juncao = juncao[['tconst','titleType','primaryTitle','startYear','averageRating','numVotes']]

#Visualizar resultado:
juncao
```

```python
#Filtro a tabela para aparecer apenas os filmes:
juncao = juncao[juncao['titleType']=='movie']

#Visualizar resultado:
juncao
```

```python
#Agrupo a tabela pelos anos de estreia, tirando a media das avaliações por ano:
juncao = juncao.groupby(by='startYear')[['averageRating']].median().copy()

#Visualizar resultado:
juncao
```

```python
#Reseto o index:
juncao = juncao.reset_index()

#Visualizar resultado:
juncao
```

```python
#Excluo a linha que me aparece um ano nulo:
juncao = juncao.drop(index=223)

#Visualizar resultado:
juncao
```

```python
#Transformo a coluna startYear para inteiro:
juncao = juncao.astype({'startYear':'int'})
```

```python
#Visualizar resultado:
juncao.dtypes
```

```python
#Crio uma variavél pegando apenas a data atual
currentDateTime = datetime.now().date()

#Depois eu extraio apenas o ano dessa data e transformo para inteiro:
year = currentDateTime.strftime('%Y')
year = int(year)

#Visualizar resultado:
year
```

```python
#Crio a variável de resposta final, filtrando apenas as datas que vão até o ano atual:
media_filemes_ano = juncao[juncao['startYear']<= year]
```
