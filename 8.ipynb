### 8. Qual o número de filmes produzidos por país?

```python
import pandas as pd
```

```python
#Importando arquivo:
regiao = pd.read_csv('Arquivos/t_akas.tsv', sep='\t')
```

```python
#Vizualizar resultado:
regiao
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/title_names.tsv', sep='\t')
```

```python
#Vizualizar resultado:
title_names
```

```python
#Juntando as tabelas:
juncao = pd.merge(title_names, regiao, left_on = 'tconst', right_on='titleId', how='left')

#Vizualizar resultado:
juncao
```

```python
#Selecionando apenas as colunas que desejo apresentar na tabela:
juncao = juncao[['tconst','titleType','primaryTitle','genres','titleId','region']]
```

```python
#Filtrando a tabela para aparecer apenas filmes:
juncao = juncao[juncao['titleType']=='movie']
```

```python
#Vizualizar resultado:
juncao
```

```python
#Criando uma nova coluna para receber os valores de gêneros pré-tratados:
juncao['generos'] = juncao.apply(lambda x: x['genres'].replace(',',' '),axis=1)
```

```python
#Separando os gêneros por colunas distintas:
sep=juncao['generos'].str.split(expand=True)

#Vizualizar resultado:
sep
```

```python
#Criando uma coluna para "tconst","titleId","region" e tranzendo os valores das respectivas colunas na tabela juncao:
sep['tconst'] = juncao.apply(lambda x: x['tconst'],axis=1)
sep['titleId']= juncao.apply(lambda x: x['titleId'],axis=1)
sep['region']= juncao.apply(lambda x: x['region'],axis=1)

#Vizualizar resultado:
sep
```

```python
#Renomeando o nome das colunas:
sep.columns= ['g1','g2','g3','tconst','titleId','region']
```

```python
#Criando uma tabela para cada coluna de gênero:
g1 = sep[['g1','tconst','titleId','region']].copy()
g2 = sep[['g2','tconst','titleId','region']].copy()
g3 = sep[['g3','tconst','titleId','region']].copy()

#Vizualizar resultado:
display(g1,g2,g3)
```

```python
#Renomeando os nomeas das colunas, para gênero:
g1.rename(columns={'g1':'genero'},inplace=True)
g2.rename(columns={'g2':'genero'},inplace=True)
g3.rename(columns={'g3':'genero'},inplace=True)
```

```python
#Fazendo o empilhamento das tabelas:
union = pd.concat([g1,g2,g3])

#Vizualizar resultado:
union
```

```python
#Analisando os gêneros distintos para excluir os que não fazem parte de gênero de filme:
union['genero'].drop_duplicates()
```

```python
#Excluindo os gêneros que não são condizentes com gêneros de filmes:
clean = union[~union['genero'].isin([r'\N','Documentary','Short','News','Talk-Show','Reality-TV','Game-Show',None])]

#Vizualizar resultado:
clean
```

```python
#Ver quantos filmes há em cada região:
clean['region'].value_counts()
```

```python
#Retirando a região "\N":
filmes_regiao = clean['region'].value_counts().drop(index=r'\N')

#Vizualizar resposta:
filmes_regiao
```
