### 5. Qual o número de filmes avaliados por gênero em relação ao ano de estreia?

```python
import pandas as pd
from datetime import datetime,date
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/title_names.tsv', sep='\t')
```

```python
#Filtrando a tabela para apresentar somente os filmes:
movies = title_names[title_names['titleType']=='movie']
```

```python
#Selecionando apenas as colunas que desejo para apresentar na tabela:
movies = movies[['tconst','titleType','primaryTitle','startYear','genres']]

#Visualizar resultado:
movies
```

```python
#Criando uma nova coluna com os generos pré-tratados:
movies['genTratados'] = movies.apply(lambda x: x['genres'].replace(',',' '),axis =1)

#Visualizar resultado:
movies
```

```python
#Separando os diferentes gêneros que copoem um filme em colunas diferentes:
sep = movies['genTratados'].str.split(expand=True)

#Visualizar resultado:
sep
```

```python
#Incluindo uma coluna "startYear" com os valores de "startYear" da tabela movies:
sep['startYear']=movies.apply(lambda x: x['startYear'],axis =1)

#Visualizar resultado:
sep
```

```python
#AQUI É SÓ UMA OBSERVAÇÃO, DE QUE TEM COMO FAZER O MESMO QUE A CÉLULA DE CIMA, DE OUTRA FORMA:
#Salvo em uma variável os valores de "startYear":
startYear = movies['startYear']

#Crio uma coluna nova "anoEstreia" com os valores de "startYear":
sep['anoEstreia'] = startYear

#Visualizar resultado:
sep
```

```python
#Apago a coluna "anoEstreia" pois era só para teste:
sep.drop(columns='anoEstreia',inplace=True)
```

```python
#Renomeio o nome das colunas:
sep.columns=['g1','g2','g3','startYear']

#Visualizar resultado:
sep
```

```python
#Crio uma tabela para cada coluna de gênero:
g1=sep[['g1','startYear']].copy()
g2=sep[['g2','startYear']].copy()
g3=sep[['g3','startYear']].copy()

#Visualizar resultado:
display(g1,g2,g3)
```

```python
#Renomeio o nome da coluna, das três tabelas que agrupam os generos, para o mesmo nome:"genero":
g1.rename(columns={'g1':'genero'},inplace=True)
g2.rename(columns={'g2':'genero'},inplace=True)
g3.rename(columns={'g3':'genero'},inplace=True)
```

```python
#Empilho eessas três tabelas:
union = pd.concat([g1,g2,g3])

#Reseto o index:
union = union.reset_index(drop=True)

#Visualizar resultado:
union
```

```python
#Adciono uma nova coluna "Count", responsável por armazeran a quantidade de filmes:
union['Count'] = 1

#Visualizar resultado:
union
```

```python
#Filtro a coluna de ano, retirando todos os \N para posteriormente transformar para int:
union = union[union['startYear']!=r'\N']
```

```python
#Transformo a coluna "startYear" para int:
union=union.astype({'startYear':'int'})
```

```python
#Crio uma variavél pegando apenas a data atual
currentDateTime = datetime.now().date()

#Depois eu extraio apenas o ano dessa data e transformo para inteiro:
year = currentDateTime.strftime('%Y')
year = int(year)

#Visualizar resultado:
year
```

```python
#Filtro para aparecer apenas os filmes que que foram estreados até a data atual:
union = union[union['startYear']<= year]

#Visualizar resultado:
union
```

```python
#Agrupo pela coluna "genero" e "startYear" fazendo a contagem:
agrupar = union.groupby(by=['genero','startYear']).count().copy()
```

```python
#Apago todos os indices que são \N:
agrupar = agrupar.drop(index=r'\N')
```

```python
#Visualizar resultado:
agrupar
```

```python
#Reseto index:
agrupar2 = agrupar.reset_index()
```

```python
#Visualizo os gêneros que me apresentam a tabela:
agrupar2['genero'].drop_duplicates()
```

```python
#Excluo os gêneros que não são considerados filmes:
resposta = agrupar2[~agrupar2['genero'].isin(['Documentary','Game-Show','News','Reality-TV','Short',
                                              'Talk-Show'])]
```

```python
#Visualizar resposta:
resposta
```
