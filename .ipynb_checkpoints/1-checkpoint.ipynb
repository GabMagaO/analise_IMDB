### 1. Quais são as categorias de filmes mais comuns no IMDB?
"Perguntar ao Cliente se ele quer os generos de filmes que mais aparecem, para saber se é isso mesmo"

```python
import pandas as pd
######### so comentário
```

```python
# lendo o aquivo title_names.
title_names = pd.read_csv('Arquivos/title_names.tsv', sep="\t")  
```

```python
title_names['titleType'].drop_duplicates() #Fiz um filtro para saber quais os tipos existentes
```

```python
#Filtrei a tabela para apresentar apenas os filmes da tabla
df_movies = title_names[title_names['titleType'] == 'movie'] 

#Visualizar tabela
df_movies
```

```python
#Criei uma Coluna nova com os valores da coluna genres sem a vírgula:
df_movies['gen_Tratados'] = df_movies.apply(lambda x: x['genres'].replace(',',' '),axis =1)
```

```python
df_movies.head()
```

```python
# Eu separei, em colunas distintas, os Generos Tratados:
separador = df_movies['gen_Tratados'].str.split(expand=True)

# Renomeei as colunas:
separador.rename(columns={0:'genres1',1:'genres2',2:'genres3'}, inplace=True)

#Criei uma nova coluna Count, para futuramente fazer a contagem:
separador['Count'] = separador.apply(lambda x: 1,axis =1)

#Visualizar o resultado:
separador
```

```python
#Criei uma tabela para cada coluna de gênero, para futuramente fazer o union:
genres1= separador[['genres1','Count']].copy()
genres2= separador[['genres2', 'Count']].copy()
genres3= separador[['genres3', 'Count']].copy()
```

```python
#Renomeei cada coluna, separadamente, de genres para generos:
genres1.rename(columns={'genres1':'genero'},inplace=True)
genres2.rename(columns={'genres2':'genero'},inplace=True)
genres3.rename(columns={'genres3':'genero'},inplace=True)

#Concatenei as tabelas de gêneros,zrando o index:
df_genres = pd.concat([genres1,genres2,genres3]).reset_index(drop=True)

#Visualizar a tabela:
df_genres
```

```python
#Agrupei a tabela por genero fazendo uma contagem:
df_genres = df_genres.groupby(by='genero').count().copy()

#Visualizar resultado:
df_genres
```

```python
#Ordenei, em decrescente, a tabela pela Contagem de quantas vezes, determinado gênero aparece resetando o index:
df_genres_sort = df_genres[['Count']].sort_values(by='Count',ascending=False).reset_index()

#Visualizar resultado:
df_genres_sort
```

```python
#Retire da tabela os gêneros que não são generos de filme:
#Negação de isin = ~
generos_mais_comuns=df_genres_sort[~df_genres_sort['genero'].isin(['Talk-Show','Game-Show','Short','Documentary',
                                                                    'Talk-Show','Reality-TV','News'])]

#Vizualizar resultado
generos_mais_comuns
```

```python
#Apagando a linha nula:
generos_mais_comuns= generos_mais_comuns.drop(index=3)

#Vizualisar resultado:(RESPOSTA)
generos_mais_comuns
```
