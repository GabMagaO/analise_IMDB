### 9. Quais são os top 15 melhores filmes?

```python
import pandas as pd
```

```python
#Importando arquivo:
title_names= pd.read_csv('Arquivos/title_names.tsv',sep='\t')
```

```python
#Importando arquivo:
avaliacoes = pd.read_csv('Arquivos/t_ratings.tsv', sep = '\t')
```

```python
#Juntando as tabelas:
juncao = pd.merge(title_names, avaliacoes, on='tconst', how='left')
```

```python
#Selecionando apenas as colunas que desejo apresentar na tabela:
juncao = juncao[['tconst','titleType','primaryTitle','genres','averageRating','numVotes']]
```

```python
#Filtrando para aparecer apenas filmes na tabela:
juncao= juncao[juncao['titleType']== 'movie']
```

```python
#Visualizar resultado:
juncao
```

```python
#Visualizar tipos:
juncao.dtypes
```

```python
#Visualizar o minimo e o máximo de votos:
juncao.describe()
```

```python
#RESPOSTA:
#Ordenar de forma decrescente a quantidade de votos e a média de avaliações:
juncao.sort_values(by=['numVotes','averageRating'],ascending = False).head(15)
```

```python
#Fazendo um exemplo, somente para visualizalçao, da ordenacao feita por media de avalições primeiro:
#Podemos perceber que as médias são mais altas mas o número de votos é absurdamente menor!
juncao.sort_values(by=['averageRating','numVotes'],ascending = False).head(20)
```
