### 2. Qual o número de títulos por gênero?
"Como assim? Qual a quantidade de generos que possui a tabela?"

```python
import pandas as pd
```

```python
#Importando arquivo:
title_names = pd.read_csv('Arquivos/t_basics.tsv', sep ='\t')
```

```python
#Filtrando para aparecer na tabela, apenas os filmes:
title_names = title_names[title_names['titleType'] == 'movie']

#Visualizar resultado:
title_names
```

```python
#Selecionando apenas as colunas que eu quero que sejam apresentadas na minha tabela:
title_names=title_names[['tconst','titleType','genres']]

#Visualizar resultado:
title_names
```

```python
#Criando uma coluna nova para tratar os dados de generos:
title_names['gen_Tratados']= title_names.apply(lambda x: x['genres'].replace(',',' '),axis=1)

#Visualizar resultado:
title_names
```

```python
#Separando os generos em colunas distintas:
sep = title_names['gen_Tratados'].str.split(expand=True)

#Visualizar resultado:
sep
```

```python
#Renomeando as colunas:
sep.columns = ['g1','g2','g3']

#Visualizar resultado:
sep
```

```python
#Criando uma variável com os valores de tconst
tconst = title_names['tconst']

#Criando uma nova coluna "tconst" incluindo os valores de tconst nela:
sep['tconst']=tconst

#Visualizar resultado:
sep
```

```python
#Criando uma tabela para cada coluna de genero:
g1=sep[['g1','tconst']].copy()
g2=sep[['g2','tconst']].copy()
g3=sep[['g3','tconst']].copy()

#Visualizar resultado:
display(g1)
display(g2)
display(g3)
```

```python
#Renoameando as colunas como iguais, para fazer o empilhamento:
g1.columns=['generos','tconst']
g2.columns=['generos','tconst']
g3.columns=['generos','tconst']
```

```python
#Empilhei as colunas por gênero, resetando o index:
union = pd.concat([g1,g2,g3]).reset_index(drop=True)

#Visualizar resultado:
union
```

```python
#Agrupei a tabela pelos gêneros contando quantos filmes existem para cada gênero:
filmes_genero = union.groupby(by='generos')['tconst'].count().copy()

#resetei o index:
filmes_genero = filmes_genero.reset_index()

#Visualizar resultado:
filmes_genero
```

```python
#Apaguei da tabelas os genêros que não são de filmes:
filmes_genero=filmes_genero[~filmes_genero['generos'].isin(['Talk-Show','Game-Show','Short','Biography',
                                                            'Talk-Show','Reality-TV','News','Documentary'])]
```

```python
#Apaguei o gênero \N que tem index = 28:
filmes_genero = filmes_genero.drop(index=28)

#Visualizar resposta:
filmes_genero
```
